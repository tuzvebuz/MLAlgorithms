## Manipulate the array to change shapes

An array is made with 3 rows and 2 columns.
That shape is reversed by the editing the shape attribute, making it 2 columns and 3 rows. Reversing the shape does not make [1,2,3] appear all in one column but rather skipping one value and taking the next one. Like [1,3,5] shown below.

```python
n1ShapeArray = np.array([[1,2,3],[4,5,6]])
print(n1ShapeArray)
n1ShapeArray.shape = (3,2)
print("\n", n1ShapeArray)
```
