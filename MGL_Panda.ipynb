## Series object

Make a series object with Series(), this takes an array so with numpy we can initialize an array from 1 to 10 with an increasement of 1.

```python
n1 = np.arange(1,10,1)
s1 = pd.Series(n1)
```

## Series object operations
Same logic but different range and increasement.
However, we also show operations between series objects. s2-s1 will remove object at index 1 in s1 from the object at index 1 in s1. This will continue for each corresponding index. This will continue untill one runs out of indexes, in that case NaN will take its place.

```python
n2 = np.arange(10,100,10)
s2 = pd.Series(n2)
s2 - s1
```
